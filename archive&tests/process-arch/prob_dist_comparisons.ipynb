#### Comparison of Data with a Trapped Particle. Ornstein-Ulhenbeck Process for a Trapped Particle
The particle should behave as a trapped particle, with a different ''temperature'' due to the interactions in the system. We do a Ornstein-Ulhenbeck Process Simulation.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Parameters of Ornstein-Ulhenbeck Process
n_runs = 1000
#tot_time = 30
#dt = 0.329
dt = 0.01
gamma = 1
k = 1
x_0 = 0
g = 1 # Related with temperature
#n_steps = int(tot_time/dt)
n_steps = 1000
```

OU Process

```python
x = np.zeros(n_runs) 
for i in range(n_runs):
    x[i] = x_0
    for i in range(n_steps):
        # i + 1
        dW = np.sqrt(dt) * np.random.randn()       
        dx = (-k / gamma) * x[i] * dt + g * dW 
        x[i] = x[i] + dx

plt.hist(x, bins=50)
plt.title("Simulacion Proceso OU")
plt.xlabel("x")
plt.ylabel("counts")
plt.show()       
```
